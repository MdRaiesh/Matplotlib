# Python Matplotlib

Matplotlib is a python library used for data visualization

We can create bar plots, scatter plots, histograms, and a lot more with matplotlib                   

## Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title("X vs Y")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=4)
plt.plot(x,y2,color='r',linestyle='-',linewidth=2)
plt.title("Two lines in same plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=4)
plt.title("Sub plots")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=2)
plt.title("Sub plots")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

## Bar Plot

```python
student = {"Bob":97,"Matt":77,"Sam":55}
names = list(student.keys())
values = list(student.values())
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# Horizontal Bar Chart

student = {"Bob":97,"Matt":77,"Sam":55}
names = list(student.keys())
values = list(student.values())
plt.barh(names,values,color='r')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

## Scatter plot

```python
import numpy as np
from matplotlib import pyplot as plt

x = [10,20,30,40,50,60,70,80,90]
m = [8,2,3,6,9,5,4,1,7]
plt.scatter(x,m,marker="*",c="g",s=100)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
m = [8,2,3,6,9,5,4,1,7]
n = [7,2,6,8,9,3,2,5,4]
plt.scatter(x,m,marker="*",c="g",s=100)
plt.scatter(x,n,marker=".",c="r",s=200)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
m = [8,2,3,6,9,5,4,1,7]
n = [7,2,6,8,9,3,2,5,4]
plt.subplot(1,2,1)
plt.scatter(x,m,marker="*",c="g",s=100)
plt.subplot(1,2,2)
plt.scatter(x,n,marker=".",c="r",s=200)
plt.show()
```

## Histogram



```python
import numpy as np
from matplotlib import pyplot as plt

data = [4,2,7,4,2,6,7,2,7,4,7,6,2,4]
plt.hist(data,color='g',bins=5)
plt.show()
```

## Box plot


```python
import numpy as np
from matplotlib import pyplot as plt

one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5]
data = list([one,two,three])
plt.boxplot(data)
plt.show()
```

## Violin plot

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5]
data = list([one,two,three])
plt.violinplot(data,showmedians=True)
plt.show()
```

## Pie Chart

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [56,89,45,76]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','grey','blue','black'])
plt.show()
```

## DoughNut Chart

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [56,89,45,76]
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
